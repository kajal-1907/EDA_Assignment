```python
# 1  What is the average rating for each product category
df['main_category'] = df['category'].str.split('|').str[0]

# Convert the 'rating' column to numeric (ignoring errors for any invalid values)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the average rating for each main category
average_ratings = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
```

```python
# 2  What are the top rating_count products by category
df['main_category'] = df['category'].str.split('|').str[0]

# Convert 'rating_count' to strings before replacing commas, then to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=True), errors='coerce')

# Get the top product by rating_count for each main category
top_products = df.loc[df.groupby('main_category')['rating_count'].idxmax()]

# Select relevant columns for clarity
top_products = top_products[['main_category', 'product_name', 'rating_count', 'rating']]
```

```python
top_products
```

```python
# 3  What is the distribution of discounted prices vs. actual prices
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', '', regex=True), errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', '', regex=True), errors='coerce')

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['actual_price'], df['discounted_price'], alpha=0.5, label='Products')
plt.plot([df['actual_price'].min(), df['actual_price'].max()], 
         [df['actual_price'].min(), df['actual_price'].max()], 
         color='red', linestyle='--', label='No Discount Line')

# Customize the plot
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Discounted Price (₹)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# 4  How does the average discount percentage vary across categories
df['main_category'] = df['category'].str.split('|').str[0]

# Convert 'discount_percentage' to string, remove '%' symbol, then to numeric
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].astype(str).str.replace('%', ''), errors='coerce')

# Calculate the average discount percentage for each category
average_discount = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)
```

```python
average_discount
```

```python
# 5  What are the most popular product names
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=True), errors='coerce')

# Find the top 10 most popular product names based on 'rating_count'
most_popular_products = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False).head(10)
```

```python
# 6  What are the most popular product keywords
from collections import Counter

all_product_names = ' '.join(df['product_name'].dropna().astype(str))

# Split into individual words (keywords)
keywords = all_product_names.split()

# Count the frequency of each keyword
keyword_counts = Counter(keywords)

# Convert to a DataFrame for sorting and display
keyword_df = pd.DataFrame(keyword_counts.items(), columns=['Keyword', 'Frequency']).sort_values(by='Frequency', ascending=False)
```

```python
# 8  What is the correlation between discounted_price and rating
df['discounted_price'] = pd.to_numeric(df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '', regex=True), errors='coerce')

# Convert 'rating' to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with missing values in discounted_price or rating
df = df.dropna(subset=['discounted_price', 'rating'])

# Calculate the correlation coefficient
correlation = df['discounted_price'].corr(df['rating'])
```

```python
# 10  Identify any potential areas for improvement or optimization based on the data analysis
df['discounted_price'] = pd.to_numeric(
    df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '', regex=True), errors='coerce')
df['actual_price'] = pd.to_numeric(
    df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '', regex=True), errors='coerce')
df['rating_count'] = pd.to_numeric(
    df['rating_count'].astype(str).str.replace(',', '', regex=True), errors='coerce')
df['discount_percentage'] = pd.to_numeric(
    df['discount_percentage'].astype(str).str.replace('%', ''), errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Extract main category
df['main_category'] = df['category'].str.split('|').str[0]

# 1. Categories with Low Average Ratings
low_rated_categories = df.groupby('main_category')['rating'].mean().sort_values().head(5)

# 2. High-Discount, Low-Rating Products
high_discount_low_rating = df[(df['discount_percentage'] > 50) & (df['rating'] < 3)]

# 3. Products with Low Review Counts
low_reviewed_products = df[df['rating_count'] < 10]

# 4. Pricing Outliers
category_avg_prices = df.groupby('main_category')['discounted_price'].mean()
df['price_deviation'] = df['discounted_price'] / df['main_category'].map(category_avg_prices)
outliers = df[(df['price_deviation'] > 2) | (df['price_deviation'] < 0.5)]
```

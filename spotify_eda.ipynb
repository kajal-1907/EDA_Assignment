```python
# 2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Popularity'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Popularity Among Tracks', fontsize=16)
plt.xlabel('Popularity', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# 3  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (ms)'], df['Popularity'], color='purple', alpha=0.6, edgecolor='k')
plt.title('Relationship Between Track Duration and Popularity', fontsize=16)
plt.xlabel('Duration (ms)', fontsize=12)
plt.ylabel('Popularity', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```

```python
# 4  Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot

artist_track_counts = df['Artist'].value_counts()

# Find the artist with the highest number of tracks
most_tracks_artist = artist_track_counts.idxmax()
most_tracks_count = artist_track_counts.max()
print(f"The artist with the highest number of tracks is '{most_tracks_artist}' with {most_tracks_count} tracks.")

# Visualize the count of tracks for each artist using a count plot
plt.figure(figsize=(12, 20))
sns.countplot(data=df, y='Artist', order=artist_track_counts.index, palette='viridis')
plt.title('Number of Tracks by Artist', fontsize=16)
plt.xlabel('Number of Tracks', fontsize=12)
plt.ylabel('Artist', fontsize=12)
plt.show()

```

```python
# 5  What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
least_popular_tracks = df.sort_values(by='Popularity', ascending=True).head(5)

```

```python
# 6  Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist

artist_avg_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

# Get the top 5 most popular artists based on average popularity
top_5_artists = artist_avg_popularity.head(5)

# Identify the artist with the highest average popularity
most_popular_artist = top_5_artists.idxmax()
highest_avg_popularity = top_5_artists.max()
```

```python
# 7  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
top_5_artists = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5).index

# Filter the dataset for the top 5 artists and find their most popular tracks
most_popular_tracks = df[df['Artist'].isin(top_5_artists)].sort_values(by=['Artist', 'Popularity'], ascending=[True, False])

# Select the most popular track for each artist
result = most_popular_tracks.groupby('Artist').first().reset_index()
```

```python
# 8  Visualize relationships between multiple numerical variables simultaneously using a pair plot
numerical_cols = ['Popularity', 'Duration (ms)']

# Create a pair plot with histograms on the diagonal
sns.pairplot(df[numerical_cols], diag_kind='hist')

# Display the plot
plt.show()
```

```python
# 9  Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot

plt.figure(figsize=(18, 12))

# Create a box plot to visualize track duration by artist
sns.boxplot(data=df, x='Artist', y='Duration (ms)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.title('Variation of Track Duration Across Artists')
plt.ylabel('Duration (ms)')
plt.xlabel('Artist')

# Display the plot
plt.show()
```

```python
# 10  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot
plt.figure(figsize=(12, 6))

# Create a swarm plot for track popularity by artist
sns.swarmplot(data=df, x='Artist', y='Popularity', size=5, alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.title('Distribution of Track Popularity Across Artists')
plt.ylabel('Popularity')
plt.xlabel('Artist')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Create a violin plot for track popularity by artist
sns.violinplot(data=df, x='Artist', y='Popularity', inner='quartile')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.title('Distribution of Track Popularity Across Artists')
plt.ylabel('Popularity')
plt.xlabel('Artist')

# Display the plot
plt.show()
```

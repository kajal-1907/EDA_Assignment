```python
# 1  What is the average selling price of cars for each dealer, and how does it compare across different dealers
df.columns = df.columns.str.strip()

# Group by 'Dealer_Name' and calculate the average selling price
average_price_per_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending=False)

```

```python
average_price_per_dealer
```

```python
# 2 Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends
price_variation_by_company = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

# Identify the company with the highest variation in prices
highest_variation_company = price_variation_by_company.idxmax()
highest_variation_value = price_variation_by_company.max()
```

```python
highest_variation_company, highest_variation_value
```

```python
# 3  What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare
price_distribution_by_transmission = df.groupby('Transmission')['Price ($)'].describe()

# Extract interquartile ranges (IQR) for each transmission type
price_distribution_by_transmission['IQR'] = (
    price_distribution_by_transmission['75%'] - price_distribution_by_transmission['25%']
)

# Display the full distribution and the IQR
price_distribution_by_transmission[['min', '25%', '50%', '75%', 'max', 'IQR']]
```

```python
# 4 What is the distribution of car prices across different regions
# Calculate the distribution of car prices across different regions
price_distribution_by_region = df.groupby('Dealer_Region')['Price ($)'].describe()

# Display the full distribution for each region
price_distribution_by_region[['min', '25%', '50%', '75%', 'max']]
```

```python
 # 6 How does the average selling price of cars vary by customer gender and annual income

average_price_by_gender = df.groupby('Gender')['Price ($)'].mean()

# Create income bins to categorize annual income levels
income_bins = [0, 25000, 50000, 100000, 500000, 1000000, float('inf')]
income_labels = ['<25K', '25K-50K', '50K-100K', '100K-500K', '500K-1M', '>1M']
df['Income Level'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels)

# Group by 'Gender' and 'Income Level' to calculate average selling price
average_price_by_gender_income = df.groupby(['Gender', 'Income Level'])['Price ($)'].mean()
```

```python
# 10  What are the top 5 car models with the highest number of sales, and how does their price distribution look

# Count the number of sales for each car model
top_models_by_sales = df['Model'].value_counts().head(5)

# Filter the dataset for the top 5 car models
top_models_data = df[df['Model'].isin(top_models_by_sales.index)]

# Group by model and calculate price distribution statistics
price_distribution_top_models = top_models_data.groupby('Model')['Price ($)'].describe()
```

```python
# 11 How does car price vary with engine size across different car colors, and which colors have the highest price variation 
# Extract relevant columns
filtered_data = df[['Engine', 'Color', 'Price ($)']]

# Group by Color and Engine to calculate mean and standard deviation of prices
price_stats = (
    filtered_data.groupby(['Color', 'Engine'])['Price ($)']
    .agg(['mean', 'std'])
    .reset_index()
    .rename(columns={'mean': 'Mean Price', 'std': 'Price Std Dev'})
)

# Calculate overall price variation for each color (max standard deviation across engine types)
color_variation = (
    price_stats.groupby('Color')['Price Std Dev']
    .max()
    .reset_index()
    .rename(columns={'Price Std Dev': 'Max Price Variation'})
    .sort_values(by='Max Price Variation', ascending=False)
)
```

```python
# 12 Is there any seasonal trend in car sales based on the date of sale
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Extract Month and Year
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group by Month to analyze seasonal trends
monthly_sales = df.groupby('Month')['Price ($)'].count().reset_index()
monthly_sales.columns = ['Month', 'Sales Volume']
```

```python
# 13  How does the car price distribution change when considering different combinations of body style and transmission type
# Extract relevant columns
filtered_data = df[['Body Style', 'Transmission', 'Price ($)']]

# Group by Body Style and Transmission for summary statistics
price_summary = (
    filtered_data.groupby(['Body Style', 'Transmission'])['Price ($)']
    .describe()
    .reset_index()
)

# Visualize the price distribution
plt.figure(figsize=(16, 8))
sns.boxplot(
    x='Body Style', 
    y='Price ($)', 
    hue='Transmission', 
    data=filtered_data, 
    palette='Set2'
)
plt.title('Car Price Distribution by Body Style and Transmission Type', fontsize=16)
plt.xlabel('Body Style')
plt.ylabel('Price ($)')
plt.legend(title='Transmission', loc='upper right')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

```

```python
# 14  What is the correlation between car price, engine size, and annual income of customers, and how do these features interact
import re

def extract_engine_size(engine):
    match = re.search(r'\d+(\.\d+)?', engine)
    return float(match.group()) if match else None

df['Engine Size'] = df['Engine'].apply(extract_engine_size)

# Filter necessary columns
correlation_data = df[['Price ($)', 'Engine Size', 'Annual Income']].dropna()

# Compute correlation matrix
correlation_matrix = correlation_data.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap', fontsize=16)
plt.show()

# Visualize interactions with scatter plots
plt.figure(figsize=(14, 6))

# Scatter plot: Price vs Engine Size
plt.subplot(1, 2, 1)
sns.scatterplot(x='Engine Size', y='Price ($)', data=correlation_data, alpha=0.7, color='blue')
plt.title('Price vs Engine Size', fontsize=14)
plt.xlabel('Engine Size')
plt.ylabel('Price ($)')

# Scatter plot: Price vs Annual Income
plt.subplot(1, 2, 2)
sns.scatterplot(x='Annual Income', y='Price ($)', data=correlation_data, alpha=0.7, color='green')
plt.title('Price vs Annual Income', fontsize=14)
plt.xlabel('Annual Income')
plt.ylabel('Price ($)')

plt.tight_layout()
plt.show()

```

```python
# 15  How does the average car price vary across different car models and engine types?
average_price = (
    df.groupby(['Model', 'Engine'])['Price ($)']
    .mean()
    .reset_index()
    .rename(columns={'Price ($)': 'Average Price'})
)

# Sort by Average Price (optional)
average_price = average_price.sort_values(by='Average Price', ascending=False)

# Display the top records
print("Top Car Models by Average Price:")
print(average_price.head())

# Visualize with a bar plot (top 20 for better clarity)
top_models = average_price.head(20)
plt.figure(figsize=(16, 8))
sns.barplot(
    x='Average Price', y='Model', hue='Engine', data=top_models, palette='Set2'
)
plt.title('Average Car Price by Model and Engine Type', fontsize=16)
plt.xlabel('Average Price ($)')
plt.ylabel('Car Model')
plt.legend(title='Engine Type', loc='upper right')
plt.show()

# Visualize with a heatmap (optional for larger sets)
pivot_table = average_price.pivot("Model", "Engine", "Average Price")
plt.figure(figsize=(14, 10))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="coolwarm", cbar=True)
plt.title('Heatmap of Average Car Price by Model and Engine Type', fontsize=16)
plt.xlabel('Engine Type')
plt.ylabel('Car Model')
plt.show()
```
